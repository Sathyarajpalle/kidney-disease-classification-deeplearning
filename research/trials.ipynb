```python
def get_product(x=int,y=int) -> int:
    return x*y
```

```python
get_product(3,4)
```

```python
get_product(3,"5")  #it still prints the answer which actually is an error
```

```python
#ensure annotation
from ensure import ensure_annotations
```

```python
#now when we use error it doesn't allow all the functions that gives bug'set

@ensure_annotations
def get_product(x=int,y=int) -> int:
    return x*y
```
